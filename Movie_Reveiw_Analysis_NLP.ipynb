```python
df = pd.read_csv('/content/datasets/Movie Review Dataset/movies.csv')
df.head()
```

```python
df=df.drop(['id','popularity','vote_count','release_date'],axis=1)
```

```python
columns_to_drop = [col for col in df.columns if col.startswith('Unnamed')]
df = df.drop(columns=columns_to_drop, axis=1)
```

```python
df
```

```python
df= df.drop('title',axis=1)
```

```python
import re
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.dtypes
```

```python
df['overview'] = df['overview'].astype(str)
```

```python
df.dtypes
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
```

```python
corpus=[]
for i in range(0,len(df)):
  review = re.sub('^[a-zA-Z1-9]',' ',df['overview'][i])
  review = review.lower()
  review = review.split()

  review = [lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
  review = ' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
words =[]
for i in range(0,len(corpus)):
  words.append(corpus[i].split())
```

```python
words
```

```python
import gensim
```

```python
model = gensim.models.Word2Vec(words,epochs=50)
```

```python
model.wv.index_to_key
```

```python
model.epochs
```

```python
model.corpus_count
```

```python
model.wv.similar_by_key('good')
```

```python
#remove OOV
def avg_word2vec(doc):
  # Convert generator expression to a list before calculating the mean
  word_vectors = [model.wv[word] for word in doc if word in model.wv.index_to_key]
  if not word_vectors:
    return np.zeros(model.vector_size)
  return np.mean(word_vectors, axis=0)
```

```python
!pip install tqdm
```

```python
from tqdm import tqdm
```

```python
X = []
for i in tqdm(range(len(words))):
  X.append(avg_word2vec(words[i]))
```

```python
len(X)
```

```python
x_new = np.array(X)
```

```python
x_new.shape
```

```python
y_new = df['vote_average'].apply(lambda x: 1 if x>6 else 0)
```

```python
y_new.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x_new,y_new,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
NBModel = GaussianNB().fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

```python
y_pred=NBModel.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```
